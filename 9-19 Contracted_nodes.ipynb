```python
import networkx as nx
```

```python
G1 = nx.cycle_graph(5)
```

```python
nx.draw(G1, with_labels=True, font_weight='bold')
```

```python
M = nx.contracted_nodes(G1, 3, 2 ) # 合并2个节点，写在前面的那个节点保留下来。
```

```python
nx.draw(M, with_labels=True, font_color= 'blue')
```

```python
N = nx.contracted_nodes(G1, 3, 2, self_loops=False) 
nx.draw(N, with_labels=True, font_color= 'blue')
```

```python
# 尝试画下有向图
```

```python
import matplotlib.pyplot as plt
```

```python
g = nx.DiGraph(directed = True) #导入有向图，注意
```

```python
g.add_nodes_from([1, 2, 3, 4, 5])
g.add_edge(1, 2)
g.add_edge(4, 2)
g.add_edge(3, 5)
g.add_edge(2, 3)
g.add_edge(5, 4)
```

```python
nx.draw(g, with_labels=True)  # 有向图的绘制，over
```

```python
g.add_edge(1, 2, weight = 4.7)
```

```python
nx.draw(g, with_labels=True)  # 有向图的绘制，over
```

```python
g.add_edges_from([(3, 4), (4, 5)], color='red')
nx.draw(g, with_labels=True)  # 有向图的绘制，over
```

```python
g.edges
```

```python
g.nodes
```

```python
g[1][2]['weight'] = 777
```

```python
nx.draw_networkx(g, with_labels=True)  # 有向图的绘制，over
```

```python
new_g = nx.contracted_nodes(g, 3, 5)
```

```python
nx.draw_networkx(new_g, with_labels=True)  # 有向图的,节点合并
```

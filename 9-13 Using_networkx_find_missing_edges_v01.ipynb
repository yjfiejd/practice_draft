```python
# 5) 找到all_edges中有，但是mst_edges中缺失的边
def find_diff(A, B):
    """
    其中A是集合的全部连通边，B是最小生成树的边
    """
    diff_edges = []
    for i in A:
        if i not in B:
            diff_edges.append(i)
    return diff_edges
```
